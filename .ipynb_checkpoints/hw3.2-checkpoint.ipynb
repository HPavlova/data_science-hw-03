```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def sierpinski(a,b,c,level):
    if level==0:
        plt.scatter([a[0],b[0],c[0]],[a[1],b[1],c[1]], alpha=0.5)
    else:
        sierpinski(a,((a[0]+b[0])/2, (a[1]+b[1])/2),((a[0]+c[0])/2,(a[1]+c[1])/2), level-1)
        sierpinski(((a[0]+b[0])/2, (a[1]+b[1])/2),b,((b[0]+c[0])/2,(b[1]+c[1])/2), level-1)
        sierpinski(((a[0]+c[0])/2, (a[1]+c[1])/2),((b[0]+c[0])/2,(b[1]+c[1])/2), c, level-1)

sierpinski((0,0),(2.5, 5),(5,0),5)
plt.title('Sierpinski triangle')
plt.subplots_adjust(top=1.5)
plt.show()
```
